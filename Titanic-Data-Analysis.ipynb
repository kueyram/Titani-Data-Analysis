```python
# Importing the librairies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Link to the dataset
link = "https://raw.githubusercontent.com/kueyram/DSC630/main/titanic.csv"

# Reading the csv into a pandas dataframe
titanic_df = pd.read_csv(link, sep=",")

# Print the list of columns
titanic_df.columns
```

```python
# Printing 10 random rows from the dataset
titanic_df.sample(n=10)
```

##### Question 2: Create a histogram or bar graph from your data.

```python
# Histogram of the age distribution of passengers
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic_df, x='Age', bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Bar graph of the number of passengers in each class
plt.figure(figsize=(10, 6))
sns.countplot(data=titanic_df, x='Pclass')
plt.title('Number of Passengers in Each Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

#####  Question 3: Create a boxplot from your data.

```python
# Boxplot to visualize the distribution of passenger ages by the different classes
plt.figure(figsize=(12, 6))
sns.boxplot(data=titanic_df, x='Pclass', y='Age', palette="Set3")
plt.title('Age distribution across passenger classes')
plt.xlabel('Passenger class')
plt.ylabel('Age')
plt.show()
```

##### Question 4: Create a bivariate plot from your data.
We will use a scatter plot to examine the relationship between the class of passenger, fare and survival status. This will help us to see how the fare varies across different passenger classes and how it relates to survival

```python
# Creating a bivariate plot for passenger class vs Fare by Survival status
plt.figure(figsize=(12, 6))
sns.stripplot(data=titanic_df, x='Pclass', y='Fare', hue='Survived', jitter=True, palette='Set1', dodge=True, alpha=0.7)
# Let's add a title to the plot
plt.title('Passenger class vs fare by survival status')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
# Adding a legend
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.show()
```

##### Question 5: Create any additional visualizations that will help to answer the question(s) you want to answer.

Survival reate by passenger class: This bar plot will show the survival rates within each passenger class.

```python
# Survival Rate by Passenger Class
sns.barplot(data=titanic_df, x='Pclass', y='Survived')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

Survival Rate by sex

```python
# Let's filter the dataset for survivors only
survivors_df = titanic_df[titanic_df['Survived'] == 1]

# Counting the number of survivors by sex
survivors_by_sex = survivors_df['Sex'].value_counts()

# Creating a bar plot for the number of survivors by Sex
sns.barplot(x=survivors_by_sex.index, y=survivors_by_sex.values)
plt.title('Number of survivors by Sex')
plt.xlabel('Sex')
plt.ylabel('Number of survivors')
plt.show()
```
